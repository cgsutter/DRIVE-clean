```python
import requests
import pandas as pd

url = "https://data.nssl.noaa.gov/thredds/catalog/WRDD/TAT/Data/ALBANY/probsrt/00.00/catalog.html?dataset=WRDD/TAT/Data/ALBANY/probsrt/00.00/20230101-000000.netcdf"
output_file = "your_file.nc"

response = requests.get(url)
with open(output_file, 'wb') as f:

    f.write(response.content)

print("Download complete.")
```

```python
import xarray as xr

ds = xr.open_dataset("your_file.nc")
print(ds)
```

```python
# Choose one variable to extract
# NetCDF files often contain multi-dimensional arrays. For storing to Parquet, you want to flatten the data into a table. Here’s an example of flattening:

var = 'temperature'  # ← replace with your variable of interest
df = ds[var].to_dataframe().reset_index()

print(df.head())

# df_all = ds.to_dataframe().reset_index()
# df_all.to_parquet("all_variables.parquet")
```

```python
import pandas as pd

parquet_file = "your_file.parquet"
df.to_parquet(parquet_file)

print(f"Saved as {parquet_file}")
```
