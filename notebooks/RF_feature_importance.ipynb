- proves Jay's point -- when we don't train the model on a downstream dataset, it learns that it can rely on the CNN output alone (bc CNN does well on data it was trained on). It doesn't need to look to other weather data bc CNN does good enough. But for cases it hasn't seen (which mimics how it would be operationally) the CNN is *not* good enough, and other data like weather data is helpful. The model learns how to incorporate weather data to augment the CNN's decision to make the model better on new data.
